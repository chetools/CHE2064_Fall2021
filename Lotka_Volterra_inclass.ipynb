```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
a,b,c,d = 1.5, 1., 3., 1.
```

```python
def rhs(t,z):
    x,y = z
    return np.array([a*x - b*x*y,
                     c*x*y - d*y])
```

```python
z_sol=solve_ivp(rhs, [0,100], [15,10],method='Radau',dense_output=True).sol
```

```python
t_plot = np.linspace(0,100,100)
pred,prey=z_sol(t_plot)
```

```python
fig= make_subplots()
fig.add_scatter(x=t_plot, y=pred, name='pred')
fig.add_scatter(x=t_plot, y=prey, name='prey')
```
