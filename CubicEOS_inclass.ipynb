```python
import sys
sys.path.insert(1, "/content/chetools")
import tools.che as che
import numpy as np
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
R=8.314
```

```python
p = che.Props('Water')
w = p.w  #accentric factor
```

```python
def cubic_eos_P(v, T, eos='SRK'):
    Tr = T/p.Tc
    s = {'vdW': 0., 'SRK': 1.}[eos]  #sigma
    e = {'vdW': 0., 'SRK': 0.}[eos]  #epsilon
    alpha = {'vdW': 1.,
             'SRK': (1 + (0.48 + 1.574*w - 0.176*w**2)*(1-Tr**0.5))**2}[eos]
    psi = {'vdW': 27/64, 'SRK': 0.42748}[eos]
    omega = {'vdW': 1/8, 'SRK': 0.08664}[eos]
    a = psi * alpha * R**2 * p.Tc**2 / p.Pc
    b = omega * R * p.Tc / p.Pc
    return R*T/(v-b) - a/((v+e*b)*(v+s*b))
```

```python
b_SRK = 0.8664*R*p.Tc/p.Pc
v = np.logspace(np.log10(1.01*b_SRK), -2,100)
fig = make_subplots()

for T in range(275, 800, 25):
    P=cubic_eos_P(v, T, eos='SRK')
    fig.add_scatter(x=v, y=P, name=f'{T}')
fig.update_xaxes(type='log')
```
