```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
# all mass units in kg, length dimension in meters, time in seconds
tend = 100
qin = 1.
cin = 1.
h0 = 1.
c0 = 0.1
alpha = 0.4
A = 1.
```

```python
def rhs(t, z):
    h,c = z
    dh = (qin - alpha*h**0.5)/A
    dc = (cin*qin/A - (c*alpha*h**0.5)/A -c*dh)/h
    return np.array([dh,
                     dc])
```

```python
sol=solve_ivp(rhs, [0,tend], np.array([h0,c0]), method='Radau',dense_output=True).sol
t_plot=np.linspace(0,tend,500)
h, c = sol(t_plot)
```

```python
fig = make_subplots(rows=1, cols=2)
fig.add_scatter(x=t_plot, y=h, row=1,col=1)
fig.add_scatter(x=t_plot, y=c , row=1,col=2)
```
