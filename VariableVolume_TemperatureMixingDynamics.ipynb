```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
# all mass units in kg, length dimension in meters, time in seconds
tend = 100
qin = 1.
Tin = 320.
h0 = 1.
T0 = 300
alpha = 0.4
A = 1.
```

```python
def rhs(t, z):
    h,T = z
    dh = (qin - alpha*h**0.5)/A
    dT = (Tin*qin/A - (T*alpha*h**0.5)/A -T*dh)/h
    return np.array([dh,
                     dT])
```

```python
sol=solve_ivp(rhs, [0,tend], np.array([h0,T0]), method='Radau',dense_output=True).sol
t_plot=np.linspace(0,tend,500)
h, T = sol(t_plot)
```

```python
fig = make_subplots(rows=1, cols=2)
fig.add_scatter(x=t_plot, y=h, row=1,col=1)
fig.add_scatter(x=t_plot, y=T , row=1,col=2)
```
